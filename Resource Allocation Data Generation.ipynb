```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_schools = 100
data = pd.DataFrame({
    "School_ID": [f"SCH_{i}" for i in range(1, n_schools + 1)],
    "School_Name": [f"School_{i}" for i in range(1, n_schools + 1)],
    "Region": np.random.choice(["North", "South", "East", "West"], n_schools),
    "Number_of_Students": np.random.randint(50, 1000, n_schools),
    "Performance_Score": np.round(np.random.uniform(30, 100, n_schools), 2),
    "Funding_Needed": np.random.randint(10000, 500000, n_schools),
    "Current_Funding": np.random.randint(5000, 300000, n_schools),
    "Resource_Gap": np.random.randint(0, 20, n_schools)
})
# Correctly add the Underserved_Students column
data["Underserved_Students"] = np.random.randint(0, 200, n_schools)

# Calculate additional columns
data["Funding_Deficit"] = data["Funding_Needed"] - data["Current_Funding"]
data["Resource_Gap_Per_Student"] = data["Resource_Gap"] / data["Number_of_Students"]

# Save to CSV
data.to_csv("school_resource_data.csv", index=False)
print("Synthetic data generated and saved as 'school_resource_data.csv'.")
```
